```python
import gym 
env = gym.make('FrozenLake-v1', desc=None, map_name="4x4", is_slippery=True)
env.reset()
state, _, _, _, _ = env.step(3)
env.s = state

print(state)

print("Action space: {}".format(env.action_space))
print("State space: {}".format(env.observation_space))
```

```python
n_step = 0
reward = 0

env.reset()
frames = []
done = False

while not done:
    action = env.action_space.sample()
    state, reward, terminated, truncated, info = env.step(action)
    done = terminated or truncated

    frames.append({
            'frame': env.render(),
            'state': state,
            'action': action,
            'reward': reward
            }
        )
    
    n_step += 1

print("Timesteps taken: %d"%n_step)
```

```python
from IPython.display import clear_output
from time import sleep

def print_frames(frames):
    for i, frame in enumerate(frames):
        clear_output(wait=True)
        print(frame['frame'])
        print("State: {}".format(frame['state']))
        print("Action: {}".format(frame['action']))
        print("Reward: {}".format(frame['reward']))
        sleep(1)

print_frames(frames)
```

```python
import numpy as np 
q_table = np.zeros((env.observation_space.n, env.action_space.n))
q_table
```

```python
import random 

alpha = 0.1
gamma = 0.7
epsilon = 1

all_steps = []
all_rewards = []

for i in range(1, 300001):
    env.reset()
    state = env.s

    done = False

    n_step = reward = 0
    
    while not done:
        if random.uniform(0, 1) < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(q_table[state])
        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated
        
        old_value = q_table[state, action]
        next_max = np.max(q_table[next_state])

        new_value = (1 - alpha) * old_value + alpha * (gamma * next_max + reward)
        q_table[state, action] = new_value

        n_step += 1
        state = next_state
        epsilon *= 0.99

    if i % 100 == 0:
        print("Episode: ", i)
```

```python
episode = 1000
total_steps = 0
total_reward = 0

for i in range (episode):
    env.reset()
    state = env.s
    done = False
    reward = n_step = 0

    while not done:
        action = np.argmax(q_table[state])
        state, reward, terminated, truncated, info = env.step(action)
        done = terminated or truncated

        n_step += 1

    total_steps += n_step
    total_reward += reward
    all_steps.append(n_step)
    all_rewards.append(reward)    

print("Average steps: ", total_steps / episode)
print("Average reward: ", total_reward / episode)
```
